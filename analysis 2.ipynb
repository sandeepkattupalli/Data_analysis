2) What is the Average price for vehicles based on the type of vehicle as well as on the type of gearbox.Explain me with both numerical and visualization analysis

```python
mean_prices = d.groupby(["vechile_type","gear_box"])["price"].mean().reset_index()


plt.figure(figsize=(10, 6))
sns.barplot(x='vechile_type', y='price', hue="gear_box",data=mean_prices)
plt.title("Average Price by Vehicle Type")
plt.xticks(rotation=45) 
plt.show()
```

What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot

```python
mean_prices = d.groupby(["vechile_type","gear_box"])["power_ps"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=mean_prices, x='vechile_type', y='power_ps', hue='gear_box')

plt.title('Average Power by Vehicle Type and Gearbox')
plt.show()
```

What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
data=d.groupby(['brand', 'vechile_type'])['price'].mean().unstack()
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, fmt=".2f", linewidths=.5)

plt.title('Average Price by Brand and Vehicle Type (Heatmap)')
plt.show()
```
